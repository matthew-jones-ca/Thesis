# Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import calendar
import plotly.graph_objects as go
import plotly.io as pio
from statsmodels.tsa.seasonal import STL
```

# Extensions

```python
# Load the rpy2 IPython extension
%load_ext rpy2.ipython
```

# Data

```python
# Load data
df = pd.read_csv('alberta_monthly_area_burned.csv')

# Output first 5 rows of dataset
df.head()
```

```python
# Convert the 'DATE' column to a datetime format
df['DATE'] = pd.to_datetime(df['DATE'],
                            format='%Y-%m')

# Add column containing year
df['YEAR'] = df['DATE'].dt.year

# Add column containing month
df['MONTH'] = df['DATE'].dt.month

# Output first 12 rows of dataset
df.head(12)
```

```python
# Transfer DataFrame from Python to R
%R -i df
```

# Modelling

## STL

```python
# Preform STL on data
stl_fit = STL(df['LOG_AREA_BURNED_HA'],
              seasonal=17, # Smoothing parameter for the cycle-subseries (choosen by user)
              period=12, # Number of observations in seasonal component (12 for monthly data)
              robust=True).fit()
```

## GAM

```python
%%R

# Load 'mcgv' package
suppressMessages(suppressWarnings(library(mgcv)))

# Fit GAM with interaction term 
gam = gam(LOG_AREA_BURNED_HA ~ te(MONTH, YEAR),
          data=df,
          method='REML')

# Output summary of fitted GAM
summary(gam)
```

```python
%%R -o pred_data_3d

# Create data frame to obtain predictions for 3d plot
pred_data_3d = expand.grid(
  MONTH = seq(min(df$MONTH), max(df$MONTH), length.out = 100),
  YEAR = seq(min(df$YEAR), max(df$YEAR), length.out = 100)
)

# Generate predictions from the GAM
pred_data_3d$PRED_LOG_AREA_BURNED_HA = predict(gam, newdata = pred_data_3d)
```

```python
%%R -o pred_data_2d

# Create data frame to obtain predictions for 2d plot
pred_data_2d = expand.grid(
  MONTH = 1:12,
  YEAR = seq(min(df$YEAR), max(df$YEAR), by=1)
)

# Generate predictions from the GAM
pred_data_2d$PRED_LOG_AREA_BURNED_HA = predict(gam, newdata = pred_data_2d)
```

# Plotting

## GAM 3D

```python
# Displays surface plot in jupyter notebook
pio.renderers.default = 'notebook'

# Get plotting data
surface_df = pred_data_3d.pivot(index='MONTH', columns='YEAR', values='PRED_LOG_AREA_BURNED_HA')
surface_df = surface_df[surface_df.columns[::-1]]
X, Y = np.meshgrid(surface_df.columns, surface_df.index)
Z = surface_df.values

# Create surface plot
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y, colorscale='Viridis')])

# Customize the surface plots layout and appearance
fig.update_layout(
    title={
        'text': 'Predicted Log Area Burned in Alberta',
        'x': 0.5
    },
    scene=dict(
        xaxis_title='Year',
        yaxis_title='Month',
        zaxis_title='Predicted Log Area Burned (Ha)',
        camera=dict(
            eye=dict(x=1.75, y=1.75, z=0.5)
        )
    ),
    autosize=False,
    width=800,
    height=600,
    margin=dict(l=10, r=10, b=10, t=30)
)

# # Display plot in notebook
# fig.show()

# Save surface plot as a PNG image
fig.write_image("GAM_surface.png")
```

## GAM 2D

```python
# Create subplots 
fig, axs = plt.subplots(3, 4, figsize=(15, 10), sharex=True, sharey=True)
axs = axs.flatten()  

# Month names
month_names = [calendar.month_abbr[i] for i in range(1, 13)]

for i in range(12):
    
    # Get plotting data for month
    month_data = pred_data_2d[pred_data_2d['MONTH'] == i+1]
    
    # Plot the monthly data 
    ax = axs[i]
    ax.plot(month_data['YEAR'], month_data['PRED_LOG_AREA_BURNED_HA'],
            linestyle='-',
            color='b')
    ax.set_title(month_names[i],
                 fontsize=20)
    
    # Increase x and y axis label tick sizes
    ax.tick_params(axis='x', labelsize=15)
    ax.tick_params(axis='y', labelsize=15)

# Set x and y labels
fig.text(0.5, 0.000001,
         'Year',
         ha='center',
         va='center',
         fontsize=20) 
fig.text(0.000001, 0.5,
         'Predicted Log Area Burned (Ha)',
         ha='center',
         va='center',
         rotation='vertical',
         fontsize=20)  

# Set the main plot title
plt.suptitle('Predicted Log-Area Burned by Month in Alberta',
             fontsize=25)

# Show plot
plt.tight_layout()
plt.show()
```

## STL

```python
# Set font style
plt.rc('font', family='serif')

# Retrieve the components
trend = stl_fit.trend
seasonal = stl_fit.seasonal
remainder = stl_fit.resid
observed = stl_fit.observed

# Create subplots
fig, axes = plt.subplots(4, 1, figsize=(12, 8), sharex=True)

# Plot each component
axes[0].plot(observed, color='k')
axes[0].set_title('STL of Log Monthly Area Burned in Alberta (Ha) (Non-Constant Seasonality)', fontsize=20)
axes[0].set_ylabel('Time Series', fontsize=18)

axes[1].plot(trend, color='k')
axes[1].set_ylabel('Trend', fontsize=18)

axes[2].plot(seasonal, color='k')
axes[2].set_ylabel('Seasonal', fontsize=18)

axes[3].plot(remainder, linestyle='none', marker='o', markersize=3, color='k')
axes[3].set_ylabel('Remainder', fontsize=18)
axes[3].tick_params(axis='x', labelsize=14)

# Make ticks face inward. Add ticks to top and right side of plot.
for i in range(0, 4):
    axes[i].tick_params(direction='in', top=True, right=True)

# Automatically adjust spacing in plot
plt.tight_layout()

# Show plot
plt.show()
```
